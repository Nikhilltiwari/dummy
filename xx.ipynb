```python
# Check if the problem exists in the DataFrame
def recommend(user_problem):
    # Remove leading and trailing whitespaces from user input
    user_problem = user_problem.strip()

    result = df[df["Problem"].str.strip() == user_problem]
    if not result.empty:
        # Display recommended solution and seriousness
        recommended_solution = result["Solution"].values[0]
        seriousness = result["Seriousness"].values[0]

        print(f"Recommended solution: {recommended_solution}")
        print(f"Seriousness: {seriousness}")
    else:
        print("Problem not found in the dataset.")
```
